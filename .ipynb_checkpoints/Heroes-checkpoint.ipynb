```python
#Import Dependencies
import pandas as pd
import numpy as np
```

```python
#Read File
CsvContents = pd.read_csv('Resources\purchase_data.csv')
CsvContents.head()
```

```python
#Display the total number of players

#Player Count
UniquePlayers = CsvContents['SN'].unique()
UniquePlayerCount = len(UniquePlayers)
UniquePlayerCountDF = pd.DataFrame({'Total Players':[UniquePlayerCount]})
UniquePlayerCountDF
```

```python
#Run basic calculations to obtain number of unique items, average price, etc.
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

#Purchasing Analysis
UniqueItems = CsvContents['Item Name'].unique()
UniqueItemCount = len(UniqueItems)
AverageItemPrice = CsvContents['Price'].mean()
UniquePurchaseIDs = CsvContents['Purchase ID'].unique()
UniquePurchaseIDCount = len(UniquePurchaseIDs)
TotalItemPurchases = CsvContents['Price'].sum()
PurchasingAnalysisDF = pd.DataFrame({
'Unique Items':[UniqueItemCount],
'Average Price':[AverageItemPrice],
'Number of Purchases':[UniquePurchaseIDCount],
'Total Revenue':[TotalItemPurchases],
})

PurchasingAnalysisDF["Average Price"]=PurchasingAnalysisDF["Average Price"].astype(float).map("${:,.2f}".format)
PurchasingAnalysisDF["Total Revenue"]=PurchasingAnalysisDF["Total Revenue"].astype(float).map("${:,.2f}".format)
PurchasingAnalysisDF
```

```python
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

#Gender Demographics
Genders = CsvContents['Gender'].unique()
#Unique Players with Genders
PlayerSnGender = CsvContents[['SN','Gender']]
#Drop Duplicates 
PlayerSnGender = PlayerSnGender.drop_duplicates()

#Stats
GenderCountM = PlayerSnGender.loc[PlayerSnGender['Gender']=='Male']['SN'].count()
GenderPercentM = GenderCountM/UniquePlayerCount*100
GenderCountF = PlayerSnGender.loc[PlayerSnGender['Gender']=='Female']['SN'].count()
GenderPercentF = GenderCountF/UniquePlayerCount*100
GenderCountO = PlayerSnGender.loc[PlayerSnGender['Gender']=='Other / Non-Disclosed']['SN'].count()
GenderPercentO = GenderCountO/UniquePlayerCount*100
GenderDemographicsDF = pd.DataFrame({
'Total Count': [GenderCountM, GenderCountF, GenderCountO],
'Percentage of Players': [GenderPercentM,GenderPercentF,GenderPercentO]},
index = ['Male','Female','Other / Non-Disclosed'])

GenderDemographicsDF['Percentage of Players']=GenderDemographicsDF['Percentage of Players'].astype(float).map('{:,.2f}%'.format)
GenderDemographicsDF
```

```python
#The below each broken by gender
#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Average Purchase Total per Person by Gender

#Gender Purchase Analysis
GenderGroupBy = CsvContents.groupby('Gender')
#Purchase Counts
PurchaseCountByGender = pd.DataFrame(GenderGroupBy['Price'].count())
#Average Purchase Price
AveragePurchasePriceByGender = pd.DataFrame(GenderGroupBy['Price'].mean())
#Total Purchase Value
TotalPurchasesByGender = pd.DataFrame(GenderGroupBy['Price'].sum())

GenderPurchasesDF = pd.DataFrame({
'Purchase Count':[float(PurchaseCountByGender.loc['Male']),float(PurchaseCountByGender.loc['Female']),float(PurchaseCountByGender.loc['Other / Non-Disclosed'])],
'Average Purchase Price':[float(AveragePurchasePriceByGender.loc['Male']),float(AveragePurchasePriceByGender.loc['Female']),float(AveragePurchasePriceByGender.loc['Other / Non-Disclosed'])],
'Total Purchase Value':[float(TotalPurchasesByGender.loc['Male']),float(TotalPurchasesByGender.loc['Female']),float(TotalPurchasesByGender.loc['Other / Non-Disclosed'])],
'Avg Total Purchase per Person':[float(TotalPurchasesByGender.loc['Male'])/GenderCountM,float(TotalPurchasesByGender.loc['Female'])/GenderCountF,float(TotalPurchasesByGender.loc['Other / Non-Disclosed'])/GenderCountO]
},
index = ['Male','Female','Other / Non-Disclosed'])

GenderPurchasesDF['Purchase Count']=GenderPurchasesDF['Purchase Count'].astype(float).map('{:,.0f}'.format)
GenderPurchasesDF['Average Purchase Price']=GenderPurchasesDF['Average Purchase Price'].astype(float).map('${:,.2f}'.format)
GenderPurchasesDF['Total Purchase Value']=GenderPurchasesDF['Total Purchase Value'].astype(float).map('${:,.2f}'.format)
GenderPurchasesDF['Avg Total Purchase per Person']=GenderPurchasesDF['Avg Total Purchase per Person'].astype(float).map('${:,.2f}'.format)

GenderPurchasesDF
```

```python
#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Average Purchase Total per Person by Age Group

#Age Demographics
#Create Bins
Bins = [0,9,14,19,24,29,34,39,999]
BinNames = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
CsvContents['Age Bin'] = pd.cut(CsvContents['Age'],Bins,labels=BinNames)
AgeGroupBy = CsvContents.groupby('Age Bin')
PlayerCountByAge = pd.DataFrame(AgeGroupBy['SN'].count())
print(PlayerCountByAge)

PurchaseCountByAge = pd.DataFrame(AgeGroupBy['Purchase ID'].count())
PurchaseCountByAgeVector = PurchaseCountByAge['Purchase ID'].tolist()

AvgPurchasePriceByAge = pd.DataFrame(AgeGroupBy['Price'].mean())
AvgPurchasePriceByAgeVector = AvgPurchasePriceByAge['Price'].tolist()

TotalPurchaseValueByAge = pd.DataFrame(AgeGroupBy['Price'].sum())
TotalPurchaseValueByAgeVector = TotalPurchaseValueByAge['Price'].tolist()

#Age Bin Demographics
#Unique Players with Age Groups
PlayerSnAgeBin = CsvContents[['SN','Age Bin']]
#Drop Duplicates 
PlayerSnAgeBin = PlayerSnAgeBin.drop_duplicates()
PlayersGroupByAge = PlayerSnAgeBin.groupby('Age Bin')
PlayerCountByAge = pd.DataFrame(PlayersGroupByAge['SN'].count())
PlayerCountByAgeVector = PlayerCountByAge['SN'].tolist()

#Find Average Total Purchase By Age Bin
AverageTotalPurchasePerPersonByAgeBin = np.divide(TotalPurchaseValueByAgeVector,PlayerCountByAgeVector)

AgePurchasesDF = pd.DataFrame({
'Purchase Count':PurchaseCountByAgeVector,
'Average Purchase Price':AvgPurchasePriceByAgeVector,
'Total Purchase Value':TotalPurchaseValueByAgeVector,
'Avg Total Purchase per Person':AverageTotalPurchasePerPersonByAgeBin
},
index = BinNames)

AgePurchasesDF['Average Purchase Price']=AgePurchasesDF['Average Purchase Price'].astype(float).map('${:,.2f}'.format)
AgePurchasesDF['Total Purchase Value']=AgePurchasesDF['Total Purchase Value'].astype(float).map('${:,.2f}'.format)
AgePurchasesDF['Avg Total Purchase per Person']=AgePurchasesDF['Avg Total Purchase per Person'].astype(float).map('${:,.2f}'.format)

AgePurchasesDF
```

```python
#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#SN
#Purchase Count
#Average Purchase Price
#Total Purchase Value

#Top 5 Spenders
SNGroupBy = CsvContents.groupby('SN')

PurchasesBySN = pd.DataFrame(SNGroupBy['Price'].sum())
PurchasesBySN = PurchasesBySN.sort_values('Price',ascending=False)
#Only Take Top 5
PurchasesBySN = PurchasesBySN.iloc[0:5]
Top5Purchasers = PurchasesBySN.index.tolist()
#Stats For Top 5
PurchaseCounts = pd.DataFrame(SNGroupBy['Price'].count())
Top5SnPurchaseCounts = PurchaseCounts.loc[Top5Purchasers]['Price'].tolist()
PurchaseMeans = pd.DataFrame(SNGroupBy['Price'].mean())
Top5SnPurchaseMeans = PurchaseMeans.loc[Top5Purchasers]['Price'].tolist()
PurchaseSums = pd.DataFrame(SNGroupBy['Price'].sum())
Top5SnPurchaseSums = PurchaseSums.loc[Top5Purchasers]['Price'].tolist()

TopSpendersDF = pd.DataFrame({
'Purchase Count':Top5SnPurchaseCounts,
'Average Purchase Price':Top5SnPurchaseMeans,
'Total Purchase Value':Top5SnPurchaseSums
},
index = Top5Purchasers)

TopSpendersDF['Average Purchase Price']=TopSpendersDF['Average Purchase Price'].astype(float).map('${:,.2f}'.format)
TopSpendersDF['Total Purchase Value']=TopSpendersDF['Total Purchase Value'].astype(float).map('${:,.2f}'.format)
TopSpendersDF.index.name = 'SN'
TopSpendersDF
```

```python
#Retrieve the Item ID, Item Name, and Item Price columns
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
#Create a summary data frame to hold the results
#Sort the purchase count column in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the summary data frame

#Top 5 Items
ItemGroupBy = CsvContents.groupby(['Item ID','Item Name'])
PurchaseCountByItem = pd.DataFrame(ItemGroupBy['Price'].count())
PurchaseCountByItem = PurchaseCountByItem.sort_values('Price',ascending=False)

#Only For Top 5
PurchaseCountByItem = PurchaseCountByItem['Price']
ItemsWanted = PurchaseCountByItem.index.tolist()
Items,ItemIDs = map(list,zip(*ItemsWanted))

#Stats For Top 5
PurchasePriceByItem = pd.DataFrame(ItemGroupBy['Price'].mean())
TopItemsPurchasePrices = PurchasePriceByItem.loc[ItemsWanted]['Price'].tolist()
PurchaseValuesByItem = pd.DataFrame(ItemGroupBy['Price'].sum())
TopItemsPurchaseValues = PurchaseValuesByItem.loc[ItemsWanted]['Price'].tolist()
PurchasePriceByItem = pd.DataFrame(ItemGroupBy['Price'].count())
TopItemsPurchaseCounts = PurchasePriceByItem.loc[ItemsWanted]['Price'].tolist()

TopItemsDF = pd.DataFrame({
'Purchase Count':TopItemsPurchaseCounts,
'Purchase Price':TopItemsPurchasePrices,
'Total Purchase Value':TopItemsPurchaseValues
},
index = [Items,ItemIDs])

TopItemsDF.head()
```

```python
#Sort the above table by total purchase value in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the data frame
TopItemsDF = TopItemsDF.sort_values(by='Total Purchase Value',ascending=False)

TopItemsDF['Purchase Price']=TopItemsDF['Purchase Price'].astype(float).map('${:,.2f}'.format)
TopItemsDF['Total Purchase Value']=TopItemsDF['Total Purchase Value'].astype(float).map('${:,.2f}'.format)
TopItemsDF.head()
```
